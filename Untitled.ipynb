```python
import pandas as pd

df = pd.read_csv('C:/Users/USER/Downloads/vgames2.csv')
df
#전처리과정
```

```python
df_null = df.isnull().sum()
print(df_null)
```

```python
df_null = df.dropna(axis=0)
df_null.isnull().sum()
```

```python
# 단위 수정
def convert_sales_to_number(sales):
    if 'K' in sales:
        return float(sales.replace('K', '')) * 1000
    elif 'M' in sales:
        return float(sales.replace('M', '')) * 1000000
    else:
        return float(sales)
    
df['NA_Sales'] = df['NA_Sales'].apply(convert_sales_to_number)
df['EU_Sales'] = df['EU_Sales'].apply(convert_sales_to_number)
df['JP_Sales'] = df['JP_Sales'].apply(convert_sales_to_number)
df['Other_Sales'] = df['Other_Sales'].apply(convert_sales_to_number)
```

```python
#연도 데이터 수정
df['Year'] = df['Year'].fillna(0)
df['Year'] = df['Year'].astype(int)
df = df[(df['Year'].astype(int) > 1980) & (df['Year'].astype(int) < 2022)]
df = df[df['Year'] != 0]
df.reset_index() # 인덱스를 재색인하려고 하면 오류가 발생해서 reset_index

# 결과 확인
print(df['Year'])
```

```python
sorted_NA = df[['Genre', 'NA_Sales']].sort_values(by='NA_Sales', ascending=False)
print(sorted_NA.head(5))
# 북미 = Action 게임 선호
```

```python
sorted_EU = df[['Genre', 'EU_Sales']].sort_values(by='EU_Sales', ascending=False)
print(sorted_EU.head(5))
# 유럽 = platform 게임 선호
```

```python
sorted_JP = df[['Genre', 'JP_Sales']].sort_values(by='JP_Sales', ascending=False)
print(sorted_JP.head(5))
# 일본 = 롤플레잉 게임 선호
```

```python
sorted_OT = df[['Genre', 'Other_Sales']].sort_values(by='Other_Sales', ascending=False)
print(sorted_OT.head(5))
# other = 롤플레잉 게임 선호
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df['Total_Sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']

df_genre = df.groupby(by=['Genre'])['Total_Sales'].sum() # 전 대륙 선호하는 게임 장르
df_genre = df_genre.reset_index()

df_genre_na = df.groupby(by=['Genre'])['NA_Sales'].sum() # 북미 선호하는 게임 장르
df_genre_na = df_genre_na.reset_index()

df_genre_eu = df.groupby(by=['Genre'])['EU_Sales'].sum() # 유럽 선호하는 게임 장르
df_genre_eu = df_genre_eu.reset_index()

df_genre_jp = df.groupby(by=['Genre'])['JP_Sales'].sum() # 일본 선호하는 게임 장르
df_genre_jp = df_genre_jp.reset_index()

df_genre_ot = df.groupby(by=['Genre'])['Other_Sales'].sum() # 그 외 선호하는 게임 장르
df_genre_ot = df_genre_ot.reset_index()

df_genre = df_genre.combine_first(df_genre_na).combine_first(df_genre_eu).combine_first(df_genre_jp).combine_first(df_genre_ot)
df_genre = df_genre.set_index("Genre")

sns.set(style='darkgrid')
plt.figure(figsize=(15, 10))
plt.title("Preferred Game Genre", fontsize=30)
plt.xlabel("Genre of game", fontsize=20)
plt.ylabel("Total sales", fontsize=20)

sns.lineplot(data=df_genre, markers=True);
```

```python
df_year = df.groupby(df['Year'].astype(int))['Name'].count().reset_index()
df_year.columns = ['Year', 'Count']

plt.figure(figsize=(15,5))
sns.barplot(x='Year', y='Count', data=df_year)
plt.xticks(rotation=90)
plt.title('Number of Video Game Releases by Year')
plt.show()

#연도 별 게임 출시 수
```

```python
#연도별 게임 판매량
yearly_sales = df.groupby('Year').sum()[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales.index, yearly_sales['NA_Sales'], label='NA Sales')
plt.plot(yearly_sales.index, yearly_sales['EU_Sales'], label='EU Sales')
plt.plot(yearly_sales.index, yearly_sales['JP_Sales'], label='JP Sales')
plt.plot(yearly_sales.index, yearly_sales['Other_Sales'], label='Other Sales')
plt.title('Global Video Game Sales by Region (1980-2020)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Sales (in millions)', fontsize=12)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# 장르 판매량

genre_sales = df.groupby('Genre').sum()[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
genre_sales = genre_sales.sort_values(by='NA_Sales', ascending=False)

# 시각화
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_sales.index, y=genre_sales['NA_Sales'], color='blue', alpha=0.5, label='NA')
sns.barplot(x=genre_sales.index, y=genre_sales['EU_Sales'], color='purple', alpha=0.5, label='EU')
sns.barplot(x=genre_sales.index, y=genre_sales['JP_Sales'], color='green', alpha=0.5, label='JP')
sns.barplot(x=genre_sales.index, y=genre_sales['Other_Sales'], color='yellow', alpha=0.5, label='OT')
plt.title('Genre Sales', fontsize=20)
plt.xlabel('Genre', fontsize=15)
plt.ylabel('Sales (million)', fontsize=15)
plt.legend(loc='upper right')
plt.show()
```

```python
platform_sales = df.groupby('Platform')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
platform_sales['Global_Sales'] = platform_sales.sum(axis=1)

sns.set_style('whitegrid')

plt.figure(figsize=(15,6))
sns.barplot(x=platform_sales.index, y=platform_sales['Global_Sales'])
plt.title('Global Platform Sales')
plt.ylabel('Sales')
plt.show()
# X360, PS, 닌텐도 순으로 플랫폼 별 게임 판매량이 많다.
```

```python
df_sales = df[['Name', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
# 개별 게임별 총 판매량 계산
df_sales.loc[:, 'Total_Sales'] = df_sales['NA_Sales'] + df_sales['EU_Sales'] + df_sales['JP_Sales'] + df_sales['Other_Sales']

# 총 판매량 상위 10개 게임 추출
df_sales_top10 = df_sales.nlargest(10, 'Total_Sales')
df_sales_top10
```

```python
# 시각화

sns.set_style('whitegrid')
plt.figure(figsize=(15,6))
sns.barplot(x='Name', y='Total_Sales', data=df_sales_top10)
plt.title('Top 10 Video Games by Total Sales')
plt.ylabel('Total Sales (in millions)')
plt.xticks(rotation=90)
plt.show()
```

```python
x360_sales_by_region = df[df['Platform']=='X360'].groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# 시각화
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(data=x360_sales_by_region, palette="tab10")
plt.title('X360 Sales by Region', fontsize=20)
plt.xlabel('Region', fontsize=15)
plt.ylabel('Sales (million)', fontsize=15)
plt.show()
```

```python
df_grouped = df[df['Year'] >= 2010].groupby(['Year', 'Genre']).size().reset_index(name='count')
# 각 연도별 장르 수의 비율 구하기
df_grouped['ratio'] = df_grouped.groupby('Year')['count'].apply(lambda x: x / float(x.sum()))
# 상위 3개 장르 추출하기
top_3_genres = df_grouped.sort_values(by='ratio', ascending=False).groupby('Year').head(3)

# 시각화하기
plt.figure(figsize=(20,5))
sns.barplot(x='Year', y='ratio', hue='Genre', data=top_3_genres)
plt.title('Top 3 Genres by Ratio of Game Releases per Year')
plt.xlabel('Year')
plt.ylabel('Ratio of Game Releases')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.show()
```
